```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import WMAP9 as cosmo
from scipy.integrate import quad
#from astropy.cosmology import Planck15   # You can choose a different cosmology if needed
import astropy.units as u
from astropy import constants as const
from astropy.cosmology import FlatLambdaCDM
from numba import jit
```

```python
B0 = 1.0  # magnetic field strength
L = 1.0   # half-width of the current sheet
P0 = 0.5  # background plasma pressure
mu0 = 1.2566370614e-6  # vacuum permeability (H/m)

# Define the spatial domain
x = np.linspace(-5, 5, 400)

# Calculate the magnetic field and pressure profiles
By = B0 * np.tanh(x / L)
P = P0 + (B0**2 / (2 * mu0)) * (1/np.cosh(x / L)**2)

# Plotting
plt.figure(figsize=(12, 6))

# Magnetic field plot
plt.subplot(1, 2, 1)
plt.plot(x, By, label='B_y(x)')
plt.xlabel('x')
plt.ylabel('Magnetic Field B_y')
plt.title('Harris Current Sheet Magnetic Field')
plt.grid(True)
plt.legend()

# Pressure plot
plt.subplot(1, 2, 2)
plt.plot(x, P, 'r', label='P(x)')
plt.xlabel('x')
plt.ylabel('Pressure P')
plt.title('Harris Current Sheet Pressure Profile')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
